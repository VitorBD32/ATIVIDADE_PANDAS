```python
# Definir uma função para remover outliers usando o IQR
def remove_outliers(teste, column):
    # Calcular o primeiro quartil (Q1) e o terceiro quartil (Q3)
    Q1 = teste[column].quantile(0.25)
    Q3 = teste[column].quantile(0.75)
    
    # Calcular o IQR (Intervalo Interquartil)
    IQR = Q3 - Q1
    
    # Definir limites para os outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Filtrar o dataframe para remover os outliers
    df_clean = teste[(teste[column] >= lower_bound) & (teste[column] <= upper_bound)]
    
    return df_clean

# Remover outliers da coluna QT_VOTOS_NOMINAIS_VALIDOS
df_clean_votos_nominais = remove_outliers(teste, 'QT_VOTOS_NOMINAIS_VALIDOS')

# Verificar o tamanho do dataframe antes e depois
print(f'Tamanho original: {teste.shape[0]} linhas')
print(f'Tamanho após remover outliers (QT_VOTOS_NOMINAIS_VALIDOS): {df_clean_votos_nominais.shape[0]} linhas')
```

```python
# Remover outliers de ambas as colunas QT_VOTOS_NOMINAIS_VALIDOS e QT_TOTAL_VOTOS_LEG_VALIDOS
df_clean = remove_outliers(teste, 'QT_VOTOS_NOMINAIS_VALIDOS')
df_clean = remove_outliers(df_clean, 'QT_TOTAL_VOTOS_LEG_VALIDOS')

# Verificar o tamanho do dataframe antes e depois
print(f'Tamanho original: {teste.shape[0]} linhas')
print(f'Tamanho após remover outliers em ambas as colunas: {df_clean.shape[0]} linhas')
```
