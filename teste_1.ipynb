```python
teste_1 = pd.read_csv ('votacao_partido_munzona_2020_PI.csv', encoding='ISO-8859-1', sep=';')
teste_1
```

```python
media_votos = teste_1['QT_VOTOS_NOMINAIS_VALIDOS'].mean()
print(f'Média dos votos nominais válidos: {media_votos}')
```

```python
desvio_padrao_votos = teste_1['QT_VOTOS_NOMINAIS_VALIDOS'].std()
print(f'Desvio padrão dos votos nominais válidos: {desvio_padrao_votos}')
```

```python
variancia_votos = teste_1['QT_VOTOS_NOMINAIS_VALIDOS'].var()
print(f'Variância amostral dos votos nominais válidos: {variancia_votos}')
```
